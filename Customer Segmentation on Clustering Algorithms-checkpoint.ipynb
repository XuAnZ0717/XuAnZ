## Feature extraction 2

```python
cust_data['Dt_Customer'] = pd.to_datetime(cust_data['Dt_Customer'])
dates = []
for i in cust_data['Dt_Customer']:
    i = i.date()
    dates.append(i)
    
print('The date record of new customers :', max(dates))
print('The date record of previous customers :', min(dates))
```

```python
cust_data['living_with'] = cust_data["Kidhome"].replace({"Married":"Partner", "Together":"Partner", "Absurd":"Alone", "Widow":"Alone", "YOLO":"Alone", "Divorced":"Alone", "Single":"Alone",})

cust_data['child'] = cust_data['Kidhome'] + cust_data['Teenhome']
```

```python
#age of customer today
cust_data['Age'] = 2022 - cust_data['Year_Birth']

#total spendings
cust_data['Spent'] = cust_data['MntWines'] + cust_data['MntFruits'] + cust_data['MntMeatProducts'] + cust_data['MntFishProducts'] + cust_data['MntSweetProducts'] + cust_data['MntGoldProds']

#family size
cust_data['Familysize'] = cust_data['living_with'].replace({'Alone': 1, 'Partner' : 2}) + cust_data['child']
```
