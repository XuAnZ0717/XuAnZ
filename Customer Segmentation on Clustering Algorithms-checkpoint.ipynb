### Feature Engineering

### Removal of zero variance variables

### As Variance Threshold can work only upon numerical data and hence we need convert the dtypes of non-integer by Ordinal Encoder function

```python
#To see the number of unique values 
cust_data.nunique(axis = 0)
```

```python
#import ordinal encoder 
from sklearn.preprocessing import OrdinalEncoder
```

```python
# Transform the data
ord_enc = OrdinalEncoder()

cust_data[["Z_CostContact", "Z_Revenue"]] = ord_enc.fit_transform(cust_data[["Z_CostContact", "Z_Revenue"]])
```

```python
#import variance threshold
from sklearn.feature_selection import VarianceThreshold
```

```python
sel = VarianceThreshold(threshold = 0)
sel.fit(cust_data[["Z_CostContact", "Z_Revenue"]])
```
